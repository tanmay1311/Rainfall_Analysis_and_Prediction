### **3. DIVIDING DATA INTO TRAINING SET AND TESTING SET**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score ,mean_squared_error ,mean_absolute_error
```

```python
y=df["ANNUAL"] #the column that needs to be predicted
X=df[["Jan-Feb","Mar-May","Jun-Sep","Oct-Dec"]] #Input data to predict
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42) #splitting the data for model training
```

## **4. TRAINING THE MODELS**

###  **Linear Regression**

```python
model=LinearRegression() #creating an instance of the LinearRegression() model
```

```python
model.fit(X_train,y_train) #Training the data
```

```python
predsLR=model.predict(X_test) #predicting the values from the test data and storing in preds
```

```python
predsLR #this is how predictions look like
```

### **Decision Tree Regressor**

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor()
```

```python
model = model.fit(X_train,y_train)
```

```python
predsDT = model.predict(X_test)
```

```python
predsDT
```

### **Random Forest Regressor**

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model=RandomForestRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
predsRF=model.predict(X_test)
```

```python
predsRF
```

### **K Nearest Neigbors**

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
model= KNeighborsRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
predsKNN =model.predict(X_test)
predsKNN#this is how predictions look like
```

## **5. COMPARISON OF THE MODELS**

```python
# Actual Value : y_test

#Predicted values by different models:

predictions ={"Linear Regression":predsLR ,"Random Forest Regressor" : predsRF  ,"KNN" :predsKNN , "Decision Tree" :predsDT}
```

```python
print("Root Mean Squared Error :")
for k,v in predictions.items():
 print("Model {} : {} ".format(k,mean_squared_error(y_test,v,squared =False)))
```

```python
print("R2 Score :")
for k,v in predictions.items():
 print("Model {} : {} ".format(k,r2_score(y_test,v,)))
```

```python
plt.figure(figsize=(20,15)) # Setting figure size
plt.ylim(-1000,2000)
plt.plot(y_test -predsDT,linestyle=' ',marker="X",label="Decision Tree",color='pink',markersize=4,alpha=0.8,aa=True)
plt.plot(y_test -predsKNN,linestyle=' ',marker= "*",label="KNN",color='green',markersize=8,alpha=0.6,aa=True)
plt.plot(y_test - predsRF,linestyle=' ',marker="s",label="Random Forest",color='blue',markersize=4,alpha=0.2,aa=True)
plt.plot(y_test - predsLR,linestyle=' ',marker="o",label="Linear Regresion",color="brown",markersize=4,alpha=0.3,aa=True)

plt.title("\nPlotting Errors\n",fontweight="bold",fontsize=15)
plt.legend()

# Error must be centered around 0 
# Decision tree has the most deviation
# Random Forest is a bit better than Decision Tree
# Linear Regression and KNN are  good models with errors near zero.
```

```python
plt.figure(figsize=(20,15))  # Setting figure size
sns.regplot(x=y_test ,y= predsLR,label="Linear Regression")
sns.regplot(x=y_test,y=predsRF,label="Random Forest")
sns.regplot(x=y_test ,y=predsKNN,marker = "+",label="KNN")
sns.regplot(x=y_test ,y=predsDT,marker="X",label="Decision Tree")
plt.title("\nLine Plot\n",fontweight="bold",fontsize=15)
plt.legend()
```
